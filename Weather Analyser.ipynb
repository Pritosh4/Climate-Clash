```python
%%sql
SELECT *
FROM kottayam;
```

## So when was the hottest day at campus?
Knowing how hot it can get at campus as it is located in central India, let's look at the hottest day in campus and if I actually lived through it not knowing it was indeed the hottest day!

```python
%%sql
SELECT datetime, MAX(tempmax) AS maximum_temp
FROM vit
GROUP BY datetime
ORDER BY maximum_temp DESC
LIMIT 1;
```

Knowing that I was on campus in May and May the 13th (Friday) being a working day, I most definitely was out and about without realising that it was the hottest day recorded that year.

## So when was the hottest day at home?
Let's look at the hottest day back home as compared to at Campus and how big would the difference have been if I was at home!

```python
%%sql
SELECT datetime, MAX(tempmax) AS maximum_temp
FROM kottayam
GROUP BY datetime
ORDER BY maximum_temp DESC
LIMIT 1;
```

As seen clearly, there's a stark difference of 6.5C between the hottest day at home and the hottest day at campus. And both of them did not even happen in the same month!

## What was the temperature when one of the two places was experiencing their hottest day?
Seeing how the hottest day at both the places happend during different months(March & May), let's look at what the temperature was at both places when one of them was experiencing their highest.

```python
%%sql
SELECT k.datetime, v.tempmax AS vit_temp, k.tempmax AS kottayam_temp
FROM kottayam AS k
LEFT JOIN vit AS v
USING(datetime)
WHERE k.datetime = (SELECT datetime
                 FROM vit
                 ORDER BY tempmax DESC
                 LIMIT 1)
```

As seen above the difference in temperature at that exact day was 12.4C which is a huge difference and is an indication of how my body had to adapt according to the time and condition of VIT.

```python
%%sql
SELECT k.datetime, k.tempmax AS kottayam_temp, v.tempmax AS vit_temp
FROM vit AS v
LEFT JOIN kottayam AS k
USING(datetime)
WHERE datetime = (SELECT datetime
                 FROM kottayam
                 ORDER BY tempmax DESC
                 LIMIT 1)
```

Though not a huge a difference as before but the 2 locations showed a slight difference of 3.5 C.

## So when was the coldest day at campus?
Knowing how the temperature can fluctuate to both extremes at campus, let's look at the coldest day in campus and if I actually lived through it as well not knowing it was indeed the coldest day!

```python
%%sql
SELECT datetime, MIN(tempmin) AS minimum_temp
FROM vit
GROUP BY datetime
ORDER BY minimum_temp 
LIMIT 1;
```

Ooh! That is cold! Luckily after sending us back home in July 2022, I only arrived back in campus in March 2023 and hence I did not have to live through this freezing temperature!

## So when was the coldest day at home?
Now that we know how low the temperature went at VIT, let's look at the coldest day back home as compared to at Campus.

```python
%%sql
SELECT datetime, MIN(tempmin) AS minimum_temp
FROM kottayam
GROUP BY datetime
ORDER BY minimum_temp DESC
LIMIT 1;
```

As compared to the single digit temperature at campus, the lowest that the temperature went back home seems so pleasant and rejuvanating.

## What was the average temperature of each month?
Let's find out the average temperature of each month to figure which months were the coldest and the hottest month in our dataset.

```python
%%sql
SELECT EXTRACT(month FROM datetime) AS month, ROUND(AVG(temp), 2) AS monthly_avg_temp_vit
FROM vit
GROUP BY month
ORDER BY month;
```

```python
%%sql
SELECT EXTRACT(month FROM datetime) AS month, ROUND(AVG(temp), 2) AS monthly_avg_temp_kottayam
FROM kottayam
GROUP BY month
ORDER BY month;
```

As seen above the coldest month at VIT was January with an average monthly temperature of 16.43C whereas that of Kottayam was 26.3C in the month of August.

The hottest month at VIT was May with an average monthly temperature of 34.45C wheres that of Kottayam was 28.84C in the month of April.
